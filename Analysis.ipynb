```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

### Data Cleaning

```python
survey = pd.read_csv("data/survey.csv")
survey.head()
```

```python
# remove extra spaces
survey["Food or Activity Recommendations"] = survey["Food or Activity Recommendations"].str.replace(r'\s+', ' ', regex=True).str.strip()

# lower case
survey["Food or Activity Recommendations"] = survey["Food or Activity Recommendations"].str.lower()

# remove special chars
survey["Food or Activity Recommendations"] = survey["Food or Activity Recommendations"].str.replace(
    r"[^a-zA-Z\s]", "", regex=True
)

# remove stop words
stopWords = set(stopwords.words("english"))
survey["Food or Activity Recommendations"] = survey["Food or Activity Recommendations"].apply(
    lambda x: ' '.join([word for word in str(x).split() if word.lower() not in stopWords]) if pd.notna(x) else x
)

survey.head()
```

### Event Aspect Popularity

```python
points = {"1st Favorite": 3, "2nd Favorite": 2, "3rd Favorite": 1}
aspect_scores = {}
favorites = survey[["1st Favorite", "2nd Favorite", "3rd Favorite"]]

for col in favorites.columns:
    for choice in favorites[col]:
        if choice not in aspect_scores:
            aspect_scores[choice] = 0
        aspect_scores[choice] += points[col]

sorted_scores = pd.DataFrame(aspect_scores.items(), columns=['Aspect', 'Total Points'])
sorted_scores = sorted_scores.sort_values(by='Total Points', ascending=False)
sorted_scores = sorted_scores.reset_index(drop=True)
sorted_scores
```
